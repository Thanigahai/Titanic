```python
import pandas as pd 
```

```python
train=pd.read_csv('train.csv')
```

```python
train.corr()
```

## Pre-processing the training data set

```python
x=train.copy(deep=True)
x['Sex'].replace(to_replace='male',value=0,inplace=True)
x['Sex'].replace(to_replace='female',value=1,inplace=True)
```

```python
x['Age'].fillna(x['Age'].mean(),inplace=True)
x.head(6)
```

```python
x['Embarked'].replace(to_replace='S',value=1,inplace=True)
x['Embarked'].replace(to_replace='C',value=2,inplace=True)
x['Embarked'].replace(to_replace='Q',value=3,inplace=True)
x['Embarked'].fillna(0,inplace=True)
x.head(6)
```

```python
x['Embarked'].value_counts()
```

```python
prediction=['Pclass','Parch','SibSp','Fare','Age','Sex','Embarked']
d=x[prediction]
e=train.Survived
```

## Splitting the data for Training and Testing

```python
from sklearn.cross_validation import train_test_split
d_train, d_test, e_train, e_test = train_test_split(d, e,test_size=0.20,random_state=1)
```

```python
d_test.head(6)
```

### Importing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(d_train, e_train)
list(zip(prediction, logreg.coef_[0]))
```

```python
y = logreg.predict(d_test)
print(y)
```

```python
from sklearn import metrics
print (metrics.accuracy_score(e_test,y))
```

```python
print("Accuracy --> ", logreg.score(d_test, e_test)*100)
```
